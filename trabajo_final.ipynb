```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
#pagina de eleccion
url = "https://www.hites.com/zapatillas/marcas-zapatillas/?gclid=Cj0KCQiAqOucBhDrARIsAPCQL1aCL8Bb3Myh-CXLPDB_Sj1UCu35KYaQO_BKgjSiKoirA93G-QNCaNEaAkgfEALw_wcB"

page = requests.get(url)

print(page)

page = BeautifulSoup(page.content, 'html.parser')
print(page.prettify())
```

```python
all_products = page.find_all(class_="col-6 col-lg-4 px-0")
print(len(all_products))

for product in all_products:
    

    name_product = product.find(class_="link product-name--bundle")
    print(name_product.string)

    #obteniendo el precio
    price_product = product.find(class_="value")

    if price_product.string is None:
        print("Producto sin oferta")
        
        price = price_product.find(text=re.compile("\$"))
        print(int(price.replace("\n","").replace(" ","").replace("Oferta","").replace("$","").replace(".","")))
    else:
        print(int(price_product.string.replace("\n","").replace(" ","").replace("Oferta","").replace("$","").replace(".","")))
```
